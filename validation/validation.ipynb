```python
from gensim.models import FastText
import pdfplumber
from gensim.utils import simple_preprocess
```

```python
combined_text = ''
with pdfplumber.open("bean-and-legumes.pdf") as pdf:
    for page in pdf.pages:
        text = page.extract_text()
        combined_text = combined_text+text+' '
```

```python
sentences = [simple_preprocess(line) for line in combined_text.split('.')]
```

```python
model = FastText(sentences, vector_size=300, window=5, min_count=1, workers=4,epochs=10)
```

```python
model.wv.most_similar('size', topn=10)
```

```python
import gensim
from gensim.models import FastText, Phrases
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords
import re
import nltk
```

```python
nltk.download("stopwords")
```

```python
combined_text = ''
with pdfplumber.open("1-s2.0-S0959652616311210-main.pdf") as pdf:
    for page in pdf.pages:
        text = page.extract_text()
        combined_text = combined_text+re.sub(r"[^\w\s]", "", text)+' '
```

```python
combined_text
```

```python
sentences = [word_tokenize(sent) for sent in combined_text.split(".")]
stop_words = set(stopwords.words("english"))
sentences = [[word for word in sent if word not in stop_words] for sent in sentences]
```

```python
bigram = Phrases(sentences, min_count=5, threshold=10)
trigram = Phrases(bigram[sentences], min_count=5, threshold=10)
```

```python
processed_sentences = [trigram[bigram[sent]] for sent in sentences]
```

```python
processed_sentences
```

```python
model = FastText(sentences=processed_sentences, vector_size=300, window=5, min_count=5, workers=4, sg=0, epochs=10)
```

```python
model.wv.most_similar('output', topn=10)
```
